```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

### C++ `get_k_a()`

```python
PM = 7.879e-09
WSGG = 8.04549e-07
RCSLW4 = 7.60157e-06
RCSLW1 = 1.52057e-06
RCSLW8 = 1.41842e-05
RCSLW2 = 3.96744e-06

yC=np.array([PM, WSGG, RCSLW1, RCSLW2, RCSLW4, RCSLW8])
xC = np.arange(6)

matplotlib.rc('font', size=14)
plt.bar(xC, yC/WSGG)
plt.grid(axis='y')
labelsC = ["PM", "WSGG 4", "RCLSW 1", "RCLSW 2", "RCSLW 4", "RCLSW 8"]
plt.xticks(xC, labels=labelsC, rotation=45)
plt.ylabel("Normalized time per call")
plt.ylim([0, 20])
plt.tight_layout()
plt.savefig('fig_getka_c++.pdf')
print('WSGG =', WSGG)
yC/WSGG
```

### Example S3 python

These appear to be python limited, not limited by the c++ libraries. The cost is basically independent of the model used or the number of gray gases.

```python
s3_times = np.array([0.609, 0.601, 0.635, 0.613, 0.607, 0.624])
xC = np.arange(6)

matplotlib.rc('font', size=14)
plt.bar(xC, s3_times/s3_times[1])
plt.grid(axis='y')
labelsC = ["PM", "WSGG 4", "RCLSW 1", "RCLSW 2", "RCSLW 4", "RCLSW 8"]
plt.xticks(xC, labels=labelsC, rotation=45)
plt.ylabel("Normalized time per call")
plt.ylim([0.95, 1.1])
plt.tight_layout()
plt.savefig('fig_exS3_py.pdf')
print('WSGG =', s3_times[1])
```
